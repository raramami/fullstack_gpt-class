```python
# (KR)  (2026.01.27)
# 영화 이름을 가지고 감독, 주요 출연진, 예산, 흥행 수익, 영화의 장르, 간단한 시놉시스 등 영화에 대한 정보로 답장하는 체인을 만드세요.
# LLM은 항상 동일한 형식을 사용하여 응답해야 하며, 이를 위해서는 원하는 출력의 예시를 LLM에 제공해야 합니다.
# 예제를 제공하려면 FewShotPromptTemplate 또는 FewShotChatMessagePromptTemplate을 사용하세요.

from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler
from langchain.prompts.few_shot import FewShotChatMessagePromptTemplate

chat = ChatOpenAI(
    model_name="gpt-3.5-turbo",
    streaming=True,
    callbacks=[
        StreamingStdOutCallbackHandler(),
    ],
)

examples = [
    {
        "title": "The Godfather",
        "answer": """
        Here is the movie information:
        Movie Title : The Godfather
        Director: Francis Ford Coppola
        Main Cast: Marlon Brando, Al Pacino, James Caan, Robert Duvall, Diane Keaton
        Budget: $6 million
        Box Office: Approx. $250-291 million
        Genre: Crime, Drama
        Synopsis: The aging patriarch of an organized crime dynasty, Vito Corleone, transfers control of his clandestine empire to his reluctant son, Michael. The film follows the violent transition of power and the moral corruption of the Corleone family.
        """,
    },
    {
        "title": "Parasite",
        "answer": """
        Here is the movie information: 
        Movie Title : Parasite
        Director: Bong Joon-ho
        Main Cast: Song Kang-ho, Lee Sun-kyun, Cho Yeo-jeong, Choi Woo-shik, Park So-dam
        Budget: $11 million
        Box Office: Approx. $263 million
        Genre: Dark Comedy, Thriller, Social Commentary
        Synopsis: The struggling, unemployed Kim family lives in a cramped semi-basement. They slowly scheme their way into the lives of the wealthy Park family by providing various luxury services. However, their symbiotic relationship is threatened when a hidden secret in the Parks' basement is revealed.
        """,
    },
    {
        "title": "The Giver",
        "answer": """
        Here is the movie information: 
        Movie Title : The Giver
        Director: Phillip Noyce
        Main Cast: Jeff Bridges, Meryl Streep, Brenton Thwaites, Alexander Skarsgård, Katie Holmes, Taylor Swift
        Budget: $25 million
        Box Office: Approx. $67 million
        Genre: Sci-Fi, Dystopian, Drama
        Synopsis: In a seemingly perfect society without war, pain, or suffering, young Jonas is chosen to inherit the role of the Receiver of Memory. As he learns about the "real" world—full of both beauty and tragedy—from the Giver, he realizes the high cost of his society's conformity and plans to escape.
        """,
    },
]

example_prompt = ChatPromptTemplate.from_messages(
    [
        ("human", "What is the movie information about {title}?"),
        ("ai","{answer}")
    ]
) 
    

few_shot_prompt = FewShotChatMessagePromptTemplate(
    example_prompt = example_prompt,
    examples = examples,
)

final_prompt = ChatPromptTemplate.from_messages(
    [
        ("system","You are a movie expert. "),
        few_shot_prompt,
        ("human","What is the movie information about {title}?"),
    ]
)

chain = final_prompt | chat

chain.invoke({"title":"Dune"})
```
