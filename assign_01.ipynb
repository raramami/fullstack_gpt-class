```python
from langchain.chat_models import ChatOpenAI
from langchain.prompts import ChatPromptTemplate
from langchain.callbacks import StreamingStdOutCallbackHandler

chat = ChatOpenAI(temperature=0.1,
    streaming=True,
    callbacks=[StreamingStdOutCallbackHandler()]
)

haiku_prompt = ChatPromptTemplate.from_messages([
    ("system","You are a specialized in writing poem about programming languages. Pls make Korean Poem with 5-7-5 formatting."),
    ("human","{language}"),
])

haiku_chain = haiku_prompt | chat 



explanation_prompt = ChatPromptTemplate.from_messages([
    ("system","You are an veteran programmer and poem critics. Pls read the poem , and kindly explain technical meaning and programming language feature about it."),
    ("human","{haiku}")
]
)


exp_chain = explanation_prompt | chat 

final_chain = {"haiku":haiku_chain} | exp_chain 

final_chain.invoke({
    "language":"python",
})
```
